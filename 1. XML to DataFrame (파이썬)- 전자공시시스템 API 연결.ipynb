# Library

```
from xml.etree.ElementTree import parse
import pandas as pd
# import json
```

# Data Load

```
tree = parse(path + "CORPCODE.xml")
root = tree.getroot()
df = []
for element in root.findall("list"):
    corp_code = element.findtext("corp_code")
    corp_name = element.findtext("corp_name")
    stock_code = element.findtext("stock_code")
    df.append({"corp_code":corp_code, "corp_name":corp_name, "stock_code":stock_code})
```

```
df = pd.DataFrame(df)
```

```
print(df.shape)
df
```
